```python
 # Histogram
data = [10,20,20,20,30,30,40,10]
# inge 10 rendu vati , 20 rendu vati, 30 moonu vati 40 oru vati vandhirki, adhe dha histogram la theridhi
plt.hist(data)
plt.show()

#Ethana vaati value repeat aagudhu nu distribution kaatum
```

```python
plt.hist(data, bins=4)
# bins na verum , athana bars mattume namma use panananum , inge bins=4 so 4 bars dha use pananum
# meaning motha data andha 4 categorory groups ah mathi , oru oru group kulla yendha data varumo andha bar varum

plt.show()


# simple ah solanumna , inge data least value 10 , highest value 40 , so 10 to 40 nadula dha data irki
# 40-10 is 30 , so difference 30 dha,, nama use panavendia categories (bins 4 dha)
# so 30/4 = 7.5 , so oru oru bin , meaning oru oru bar 7.5 width la varum x axis la
# next example pakalam
```

```python
scores = [5, 6, 4, 1, 5, 7, 4, 9, 6, 7]
# inge pakalam , maximum value 9, minimum value 1, so 9-1 is 8 , so motha 8/bins
# 8 divided by 4 bins i.e 8/4 is 2
# so oru oru bin width 2 irukum , ipidi motha data 4 bins kulla divide ayindhirkum
# diagram pathingana first bin 1 landhu 3 varkum iriki meaning (3-1) is 2 width
# min 1 ,max 9 , so 1+2 is 3 ,so 1 to 3 oru bin (1 ku 3 ku nadula irukara data lam inge irukum)
# next 3 + 2(bin width) is 5 , so 3 to 5 oru bin (3 to 5 ku nadula 4 mattum 2 times irki , so count 2)
# 3 to 5 means , considering 3, 4 not considering 5(upperbound)
plt.hist(scores, bins=4)
plt.xlabel('Score')
plt.ylabel('Count')
plt.show()
```

```python


scores = [45, 67, 34, 91, 55, 78, 40, 99, 63, 71]
# inge pakalam , maximum value 99, minimum value 34, so 99-34 is 65 , so motha 65/bins
# 65 divided by 5 bins i.e 65/5 is 13
# so oru oru bin width 13 irukum , ipidi motha data 5 bins kulla divide ayindhirkum

plt.hist(scores, bins=5)
plt.xlabel('Score')
plt.ylabel('Count')
plt.show()
```

Understanding Bins:

bins=5 → Groups data into 5 buckets (broad view)

bins=10 → Groups data into 10 buckets (balanced - recommended)

bins=20 → Groups data into 20 buckets (detailed view)

bins=[0, 50, 70, 90, 100] → Custom ranges (like grade boundaries)

```python
scores = [45, 67, 82, 91, 55, 78, 88, 92, 63, 71]

plt.hist(scores, bins=5)
plt.xlabel('Score')
plt.ylabel('Count')
plt.show()

```
